```python
# final model
fpath_final_model = './OUTPUT_FILES/final_model.h5'
fpath_true_model = "./test_model_init_100-100-100.h5"

import h5py

# read vel, xi, eta dataset 
with h5py.File(fpath_final_model, 'r') as f:
    vel = f['vel'][:]

# read true model
with h5py.File(fpath_true_model, 'r') as f:
    vel_true = f['vel'][:]

# plot
import matplotlib.pyplot as plt
import numpy as np

# plot vel
plt.figure()
plt.imshow(vel[:,:,5], origin='lower', aspect='auto')
plt.colorbar()
plt.title('vel')

# plot vel_true
plt.figure()
plt.imshow(vel_true[:,:,5], origin='lower', aspect='auto')
plt.colorbar()
plt.title('vel_init')
```
