# Plot sources and receivers

## Read src and rec files

Sources and receivers are stored in the `1_src_rec_files/src_rec_config.dat`. This file can be easily read using the PyTomoATT.

```python
from pytomoatt.src_rec import SrcRec

sr = SrcRec.read('./1_src_rec_files/src_rec_config.dat')
```

## Plot sources and receivers

All sources and receivers are saved in `sr.sources` and `sr.receivers` respectively as `pandas.DataFrame`. We can plot them using Matplotlib or PyGMT. Here we use Matplotlib to plot them.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(sr.sources['evlo'], sr.sources['evla'], 'C0.', markersize=3, label='Sources')
ax.plot(sr.receivers['stlo'], sr.receivers['stla'], 'C1^', markersize=10, label='Receivers')
ax.set_aspect('equal')
```
